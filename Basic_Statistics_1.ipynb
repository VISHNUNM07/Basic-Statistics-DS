# MEAN

# MEDIAN
**The median is another measure of central tendency that represents the middle value in a set of numbers when the numbers are arranged in order. It divides the dataset into two equal halves.**

# MODE
**The mode is another measure of central tendency that refers to the value or values that appear most frequently in a dataset. It helps identify the most common value in a set of numbers.**

```python
print(" MODE of the numerical columns")
Data.describe().mode()
```

# STANDARD DEVIATION
**Standard deviation is a measure of how spread out or dispersed the values in a dataset are from the mean (average). It tells you, on average, how much the numbers in your data differ from the mean of the dataset.**




```python
print("STANDARD DEVIATION of the numerical columns")
Data.describe().std()
```

# **Standardization of Numerical Variables**
**Standardization, also known as z-score normalization, is a technique used in statistics and data analysis to transform data such that it has a mean of 0 and a standard deviation of 1. This process allows data from different scales to be compared directly by making the distribution of the data more consistent.**

```python
Data
```

```python
numerical_colums = Data.select_dtypes( include = ['int','float'])
```

```python
Mean = Data.describe().mean()
std = Data.describe().std()

for column in numerical_columns:
  Data[column] = (Data[column] - Mean[column]) / std[column]

print("Standardized DataFrame:")
print(Data)
```

```python
for col in numerical_colums:
  plt.figure(figsize=(10, 6))
  plt.subplot(1, 2, 1)
  Data[col].plot(kind='hist', bins=50)
  plt.title('Before Standardization')
  plt.xlabel(col)
  plt.ylabel('Frequency')

  plt.subplot(1, 2, 2)
  Data[col].plot(kind='hist',bins=50)
  plt.title('After Standardization')
  plt.xlabel(col+'standardized')
  plt.ylabel('Frequency')
  plt.show()
```

# **Conversion of Categorical Data into Dummy Variables**


One-hot encoding is a technique to convert categorical data into binary vectors, where each category is represented by a binary column (0 or 1). This is necessary because many machine learning algorithms cannot handle categorical data directly.

```python
# Apply one-hot encoding to categorical columns
Data_onehot = pd.get_dummies(Data, columns=catagorical_columns)
Data_onehot
```

```python
# Create a new DataFrame with the standardized numerical columns and the one-hot encoded categorical columns
Data_transformed = pd.concat([Data_onehot.drop(columns=numerical_columns)], axis=1)
Data_transformed
```
